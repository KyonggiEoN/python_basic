# **문자열이란?**

파이썬에서 문자열은 텍스트 데이터를 다루는 데 사용되는 중요한 데이터 타입입니다. 문자열은 작은 따옴표(')나 큰 따옴표("), 세 개의 따옴표(''', """)로 둘러싸인 텍스트로 표현된다.

```python
str1 = '크롸롸라롸롸 투명드래곤이 울부지져따'
print(str1)
str2 = "크허어어엉! 백호가 우렀따"
print(str2,"\n")
str3 = '''크로롸롸롸라ㅗ뢀 투명드래곤이 백호를 물었다!
크호아아앙, 백호가 고통에 울었다
크롸뢀롸롸라ㅗ라ㅗ롸 투명드래곤이 승리의 포호를 짓엊따!'''
print(str3)
```

>파이썬의 문자열은 유니코드(Unicode) 문자의 시퀀스로 구성되며, 문자열 값은 메모리에 저장되고 해당 값은 불변(immutable)한 특성을 가진다.

1. 문자열 리터럴과 메모리 할당: 파이썬에서 문자열 리터럴(값)을 생성하면, 이러한 값은 메모리에 저장. 변수에 할당할 때는 해당 값의 참조(reference)가 변수에 저장

2. 공유 문자열 (String Interning): 작은 크기의 문자열 리터럴은 메모리에 단일로 문자열을 저장하고 여러 변수가 해당 값을 참조하게 합니다. 이로써 메모리를 절약함

3. 문자열 연결과 재할당: 문자열은 불변한 데이터 타입이므로 문자열 연결(+ 연산자)이나 문자열 재할당 시 새로운 문자열 객체 생성. 이전 문자열은 변경되지 않음

## **문자열 함수와 연산자**


```python
text = "Hi"
print(len(text)) #텍스트 갈이 확인 len()
name = "나리"
massage1 = f'{text}, {name}'
print(text + name) #텍스트 합치기
print(massage1)
print(text*20) #텍스트 반복하기
```

```python
str = '문자열 내부의 문자에 인덱스를 사용하여 접근하거나 슬라이스할 수 있습니다.'
print(str)

# 인덱싱
print(str[12])
print(str[13])
print(str[-10]) #음수 값은 문자열을 거꾸로 셈
print(str[-1]) #맨마지막 문자는 -1

# 슬라이싱
print(str[0:6])
print(str[8:16]) #8에서 시작해서 16 직전까지(16 포함 X)
print(str[:16])
print(f"{_:<13}", str[8:])
print(str[0:14:2]) #0부터 14직전까지 2씩 증가(default 1)
print(str[14:0:-2]) #거꾸로도 가능
print(str[len(str):0:-1]) #-1하면 문자열 반전 가능(주의, 첫글자 소실)
print(str[::-1]) #문자열 반전
a = 6
print(str[0:a]) #변수로 슬라이싱이나 인덱싱 가능
```

## **문자열 메서드**

문자열 메서드는 문자열 객체에 적용되는 함수로, 문자열을 조작하거나 변환하는 다양한 작업을 수행할 수 있게 해줍니다.

```python
text = "Hello, Python!"
print(text.upper())
print(text.lower())

#str.count(str, __start, __end)
print(text.count('l'))
print(text.count('ll'))

#str.find(str, __start, __end)
print('find 메서드') #특정 부분 문자열의 첫 번째 등장 인덱스 확인
print(text.find('l'))
print(text.find('ll'))
print(text.find('z')) #없으면 -1 반환
print(text.rfind('l')) #뒤에서부터 찾아라
print(text.find('o', 5)) #5번 문자부터 시작하라
print(text.find('o', text.find('o')+1))
```

```python
text = "Hello, Python!"
new_text = text.replace("Python", "World") #Python을 찾아서 World로 교체하라
print(new_text)

text2 = "   Hello, Python!   "
print(text2)
print(text2.strip())
print(text2.lstrip()) #왼쪽 공백 제거
print(text2.rstrip()) #오른쪽 공백 제거

text = '김사과 반하나 오렌지 이메론'
names = text.split() # 공백을 기준으로 분리 -> 리스트로 저장
print(names)

text = '김사과,반하나,오렌지,이메론'
names = text.split(',') # ','을 기분으로 분리 -> 리스트로 저장
print(names)

text = 'Hello'
print('아'.join(text)) #문자열 사이에 지정된 문자 삽입

text = "Hello, Python!"
starts_with_hello = text.startswith("Hello") #Hello로 시작하나요?
ends_with_world = text.endswith("World!") #World!로 끝나나요?

print(starts_with_hello) #True
print(ends_with_world) #False
```

## **문자열은 불변 데이터 타입이다**

문자열 객체가 생성된 후 값을 변경할 수 없다.(Immutable)

한 번 생성된 문자열 내 특정 문자만 바꿀 수 없다. 일부를 변경하고 싶으면 새로운 문자열을 만들어서 교체해야 합니다.



```python
my_string = "Hello"
my_string[0] = "h"  # 오류 발생!
#TypeError: 'str' object does not support item assignment

my_string = "Hello"
new_string = "h" + my_string[1:]  # 'Hello'의 첫 글자를 바꾸기 위해 새로운 문자열 생성
print(new_string)  # 출력: "hello"
```

string은 연산하면 메모리 사용도를 급격히 높이므로 가능하다면 새로 만들어서 사용할 것
