# **조건문**

조건문은 코드의 실행 흐름을 제어하기 위해 사용하는 구문으로

주어진 조건이 참인지 거짓인지에 따라 다른 코드를 실행하도록 한다.

## **1. if문**

가장 기본적인 조건문

특정 조건이 참(Boolean_True)일 경우에만 해당 블록 내의 코드 실행

    if 조건:
      실행할 코드

여기에서 조건은 True 또는 False로 평가될 수 있는 표현식이어야 함.

조건이 True로 평가되면 if문 바로 아래의 코드 블록이 실행되고

False 일 경우, 그 블록을 건너뜀.

```python
if True:    #Tab을 이용해 블록을 구분함
    print('결과는 참입니다')
print('프로그램을 종료합니다')
```

```python
if '안녕하세요':    #Boolean에 따라 True로 평가되면 전부 True로
    print('결과는 참입니다')
print('프로그램을 종료합니다')
```

```python
if '':    #비어있는 문자열은 False
    print('결과는 참입니다')
print('프로그램을 종료합니다')
```

```python
if '안녕하세요':print('결과는 참입니다')
print('프로그램을 종료합니다')
#바로 이어쓰면 딱 하나의 함수만 사용할 수 있다
```

```python
age = int(input('나이를 입력하세요: '))

if age > 19:
    print('성인입니다')

print('입력한 나이: ', age)
```

## **2. else문**

else 문은 조건문(if, elif)에서 특정 조건이 충족되지 않을 때 실행할 코드 블록을 정의 할 때 사용함.

    if 조건1:
      코드 블록1
    else:
      코드 블록3

```python
age = int(input('나이를 입력하세요'))

if age > 19:
    print('성인입니다')
else:
    print('미성년입니다')

print('입력한 나이: ', age)
```

```python
age = int(input('나이를 입력하세요 '))

if age > 19:
    print('성인입니다')
if not age > 19: #혹시 if not은 else와 같은가?
    print('미성년입니다')

print('입력한 나이: ', age)
```

## **3. elif 문**

elif는 if- else 구조에서 추가적인 조건을 제시하기 위해 사용함.

elif는 "else if"의 줄임말.

주어진 여러 조건 중에서 하나만 참일 때 실행되는 코드 블록을 정의한다.
___
특징

  1. if문 없이 단독으로 사용 불가

  2. 필요에 따라 여러 번 사용 가능

  3. else는 선택적. 사용 할 땐 항상 if나 elif 뒤에 써야함.


    if 조건1:
      코드 블록1
    elif 조건2:
      코드 블록2
    elif 조건3:
      코드 블록3
    else:
      코드 블록4

```python
age = int(input('나이를 입력하세요 '))

if age > 19:
    print('성인입니다')
elif age > 14 and age <= 19: #아! 어차피 if에서 내려오는거라 19 이하는 필요없다
    print('청소년입니다')
else:
    print('미성년입니다')

print('입력한 나이: ', age)
```

```python
age = int(input('나이를 입력하세요 '))

if age > 19:
    print('성인입니다')
if (not age > 19) and (age > 14 and age <= 19):
    print('청소년입니다')
if not(age > 19) and not(age > 14 and age <= 19):
    print('미성년입니다')

print('입력한 나이: ', age)

#아하! if not은 elif와 else와 다르게 if구절에서 여러 조건을 여러번 조회하는구나.
```

```python
age = int(input('나이를 입력하세요: '))

if age > 19:
    print('성인입니다')
elif age > 14:
    print('청소년입니다')
elif age > 6:
    print('어린이입니다')
elif age > 0:
    print('유아입니다')
else:
    print('입력을 확인하세요')

print('입력한 나이: ', age)
```

### **문제 3**

간단한 MBTI 유형 검사를 만들어보자.

```python
print("MBTI 간단 성격 유형 검사에 오신 걸 환영합니다!")

# 첫 번째 질문: 외향/내향
answer1 = input("Q1. 새로운 사람들과 만나는 걸 좋아하시나요? (y/n): ")
if answer1.lower() == 'y':
    # 외향형으로 진행
    print("외향형으로 분석 중...")
    answer2 = input("Q2. 주변 환경에 대해 자주 인지하고 있나요? (y/n): ")

    if answer2.lower() == 'y':
        # 감각형(외향, 감각)
        answer3 = input("Q3. 계획을 세우는 것을 좋아하시나요? (y/n): ")
        if answer3.lower() == 'y':
            print("당신의 성격 유형은 ESTJ입니다!")
        else:
            print("당신의 성격 유형은 ESFP입니다!")
    else:
        # 직관형(외향, 직관)
        answer3 = input("Q3. 큰 그림을 보는 것을 좋아하시나요? (y/n): ")
        if answer3.lower() == 'y':
            print("당신의 성격 유형은 ENFP입니다!")
        else:
            print("당신의 성격 유형은 ENTJ입니다!")

else:
    # 내향형으로 진행
    print("내향형으로 분석 중...")
    answer2 = input("Q2. 구체적인 정보나 사실을 중요하게 생각하시나요? (y/n): ")

    if answer2.lower() == 'y':
        # 감각형(내향, 감각)
        answer3 = input("Q3. 일정을 엄격하게 관리하는 것을 좋아하시나요? (y/n): ")
        if answer3.lower() == 'y':
            print("당신의 성격 유형은 ISTJ입니다!")
        else:
            print("당신의 성격 유형은 ISFP입니다!")
    else:
        # 직관형(내향, 직관)
        answer3 = input("Q3. 논리적인 문제 해결을 좋아하시나요? (y/n): ")
        if answer3.lower() == 'y':
            print("당신의 성격 유형은 INTP입니다!")
        else:
            print("당신의 성격 유형은 INFJ입니다!")
```

## **5. 조건부 표현식**


조건부 표현식(Conditional Expression)은 삼항연산자로도 불림.

간단한 if-else 구문을 더 간결하게 표현하는 방법 제공.

단, 너무 복잡한 조건이나 연산 포함시 코드의 가독성이 떨어짐

따라서 적절한 상황에서만 사용할 것

      조건이 참인 경우 값 if 조건 else 조건이 거짓인 경우 값

```python
# if~else
num = int(input('숫자를 입력하세요: '))
if num % 2 == 0:
    print('짝수')
else:
    print('홀수')


# 조건부 표현식
num = int(input('숫자를 입력하세요: '))
print('짝수' if num % 2 == 0 else '홀수' if num % 2 == 1 else '0')

# C 에서 쓰던 삼항 연산자는 구동이 안 된다.
#num1 = True == True ? '참' : '거짓'
```

```python
# if~elif~else로 작성
age = int(input('나이를 입력하세요'))

if age > 19:
    print('성인입니다')
elif age > 14:
    print('청소년입니다')
elif age > 6:
    print('어린이입니다')
elif age > 0:
    print('유아입니다')
else:
    print('다시 입력해주세요')
print('입력한 나이: ', age)

# 조건부 표현식으로 작성
# 백슬래시(\)를 사용하면 줄바꿈 했을때 이어갈 수 있다! (혹은 소괄호를 이용하자.)
age = int(input('나이를 입력하세요'))
print('성인입니다') if age > 19 else print('청소년입니다') \
    if age > 14 else print('어린이입니다') if age > 6      \
    else print('유아입니다') if age > 0 else print('다시 입력해주세요')
```
