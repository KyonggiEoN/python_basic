# **1. 클로저**

내부 함수가 자신을 감싸고 있는 외부 함수의 변수에 접근 할 수 있는 특성을 가짐

외부 함수가 호출 된 후에도 그 변수의 상태를 기억, 사용할 수 있는 기능.

데이터 은닉과 상태 유지에 활용.

```python
def mul2(n):
  return n*2

print(mul2(4))

def mul5(n):
  return n*5

print(mul5)
##이대로 가다간 mul3.. mul20... mul100까지 만들겠어!!!
```

```python
# 클래스를 이용해서 만드는 mul 함수들

class Mul:
  def __init__(self, m):
    self.m = m
  def mul(self, n):
    return self.m * n

mul2 = Mul(2)
print(mul2.mul(10))
print(mul2.mul(5))

mul5 = Mul(5)
print(mul5.mul(10))
print(mul5.mul(5))
```

```python
class Mul:
    def __init__(self, m):
        print('생성자 호출')
        self.m = m
    def __call__(self, n):
      #오버라이딩 시, 객체를 호출하면 함수처럼 쓸 수 있다.
        print('call 호출')
        return self.m * n

mul2 = Mul(2)
#print(mul2.mul(10)) -> 간략하게 바뀜!
print(mul2(10))

mul5 = Mul(5)
print(mul5(10))
```

```python
# 클로저 사용하기
def mul(m):
    def wrapper(n):
        return m * n
    return wrapper

mul2 = mul(2)
print(mul2)
# mul2를 호출하면 mul(2)가 실행되고, 이것이 wrapper를 반환하므로
# 결과적으로 wrapper(n)을 호출한다!
print(mul2(10))

mul5 = mul(5)
print(mul5(10))
```

```python
def deter(N):
  def ascendent(n):
    return n * N
  def adult(n):
    return n ** N
  if N > 19:
    return adult
  elif N <= 19:
    return ascendent

humanA = deter(10)
print(humanA(10))

# 조건을 만들어서 클로저를 다채롭게 만들기 가능
```

# **2. 데코레이터**



기존 함수를 수정하거나 확장하는 함수

다른 함수를 인자로 받아 새로운 함수를 반환하여 원래 함수의 기능을 변경하거나 추가하는 데 사용됨.

주로 코드의 재사용성과 가독성을 높이는 데 활용.

```python
import time
```

```python
def func1(a, b):
  start = time.time()
  print('함수 시작')
  result = a + b

  end = time.time()
  print(f'수행 시간: {end - start}')
  return result

result = func1(10, 3)
print(result)

def func2(a, b):
  start = time.time()
  print('함수 시작')
  result = a * b

  end = time.time()
  print(f'수행 시간: {end - start}')
  return result

result = func2(10, 3)
print(result)
```

**데코레이터 만들기**

```python
import time

def func1(a, b):
  return a+b

def func2(a, b):
  return a*b

def elapsed(func):
  def wrapper(a, b):
    start = time.time()

    #변수 대신에 함수를 받아서 처리한다.
    result = func(a, b)

    end = time.time()
    print(f'수행 시간: {end - start}')
    return result

  return wrapper

'''
deco1 = elapsed(func1)
result = deco1(10, 3)
print(result)

deco2 = elapsed(func2)
result = deco2(10, 3)
print(result)
'''
@elapsed
def func1(a, b):
  return a + b

result = func1(10,3)
print(result)

# @ 의미, 추상적으로 아래 함수를 데코레이터의 인자로 받겠다?
@elapsed
def func2(a, b):
    result = a * b
    return result

result = func2(10,3)
print(result)
```
