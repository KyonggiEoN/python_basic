# 1. 산술 연산자
산술 연산자는 주로 수치 데이터 유형에 대한 기본 산술 연산을 수행하기 위해 사용됩니다.

.+ : 더하기

.- : 빼기

.* : 곱하기

/ : 나누기

//: 몫 (정수 나누기)

% : 나머지

**: 지수 (거듭제곱)

```python
print(10 + 3)
print(10 - 3)
print(10 * 3)
print(10 / 3)
#무한 소수점은 16번째 자리가 이상하게 표현됨.
# 왜? Double 자료형, 2진법 계산의 한계
print(10 // 3)
print(10 % 3)
print(10 ** 3)
```

# 2. 비교 연산자

비교 연산자는 주로 두 값을 비교하는 데 사용되며, 그 결과는 항상 Boolean(True 또는 False) 값입니다.

==: 동등

!=: 동등하지 않음

< : 작다

<=: 작거나 같다

.> : 크다

.>=: 크거나 같다

```python
print(10 > 5)
print(10 < 5)
print(10 >= 5)
print(10 <= 5)
print(10 == 5)
print(10 != 5)
```

# 3. 할당 연산자
변수에 값을 할당하는데 사용됩니다. 파이썬에서는 기본 할당 연산자 외에도 복합 할당 연산자를 제공하여 코드를 간결하게 작성할 수 있게 도와줍니다.

= : 할당

+= : 더하기 후 할당

-= : 빼기 후 할당

*= : 곱하기 후 할당

/= : 나누기 후 할당

//=: 몫 후 할당

%= : 나머지 후 할당

**=: 지수 후 할당

```python
num = 10

num += 1    # num = num + 1
print(num)
num -= 1
print(num)
num *= 10
print(num)
num /= 10
print(num)
num %= 2
print(num)
```

# 4. 비트 연산자
비트 연산자는 정수를 이진 비트로 표현했을 때 사용하는 연산자들입니다. 각 연산자는 정수의 비트 단위로 동작합니다.

& : 비트 단위 AND

.| : 비트 단위 OR

^ : 비트 단위 XOR

~ : 비트 단위 NOT

.<< : 왼쪽 시프트

.>> : 오른쪽 시프트

```python
#128 64 32 16 8 4 2 1

a = 60  # 111100
b = 13  # 001101
print('{:08b}\n{:08b}\n'.format(a, b))
#AND
#비트가 모두 1 이면 결과로 1을 반환하는 연산(하나라도 0이면 0)
print(a & b)  # 12, 즉 1100
print('{:08b}\n'.format(a & b))
#OR
#비트 둘 중 하나라도 1이면 1을 반환하는 연산(둘 다 0이면 0)
print(a | b)  # 61, 즉 111101
print('{:08b}\n'.format(a | b))
#XOR
#비트 둘 중 하나만 1이여야 1을 반환하는 연산(둘이 같으면 0)
print(a ^ b)  # 49, 즉 110001
print('{:08b}\n'.format(a ^ b))
#NOT
#비트가 1이면 0, 0이면 1을 반환 (음수일 경우 2의 보수를 구함)
print(~a)  # -61, 이 결과는 2의 보수 방식 때문입니다.
print('{:08b}\n'.format(~a))
#Lshift
#비트를 왼쪽으로 쉬프트(맨오른쪽에 0을 두 개 더 찍어라)
print(a << 2)  # 240, 즉 11110000
print('{:08b}\n'.format(a << 2))
#Rshift
#비트르르 오른쪽으로 쉬프트(맨왼쪽에 0을 두 개 더 찍어라)
print(a >> 2)  # 15, 즉 1111
print('{:08b}\n'.format(a >> 2))
```

**보수란?**

a - b = a + (-b)

-(a + 1) 을 보수라 한다.

왜? 2진법으로 10진수 표현 시 비트 자리수에 따라서 양수는 그대로지만 음수는 하나 더 높게 나오기 때문이다

(0을 포함하지 않기 때문에)

따라서, -를 붙인 후 양수 a+1을 하면 표현하려던 음수값이 나온다.

**※ 1.1 + 0.1 == 1.2가 False로 나오는 이유?**

다운로드.png

부동소수점(Floating Point) 연산의 오차 때문입니다. 컴퓨터는 10진수를 2진수로 변환하여 계산하는데, 1.1과 0.1은 2진수로 정확히 표현되지 않아서 미세한 오차가 발생합니다.

1.1과 0.1 같은 10진수를 이진수로 변환할 때, 컴퓨터는 이를 무한 소수로 처리합니다.

이진수 표현에서는 특정 소수를 정확히 표현하지 못하고 근사값으로 저장하게 됩니다.

따라서 1.1 + 0.1의 결과는 1.2에 아주 가까운 값이지만, 정확히 1.2와는 다르기 때문에 False가 반환됩니다.

# **6. 논리 연산자**

주로 불린(Boolean) 값을 조작하기 위해 사용됨.

다만, 파이썬 논리 연산자는 다른 언어와 달리 연산 결과로 Boolean을 다루는 것은 아님.

--> 마지막으로 평가된 피연산자의 값을 반환함!!!

__________
and: 논리 AND

둘 다 True여야 True

__________
or : 논리 OR

둘 중에 하나만 True여도 True

__________
not: 논리 NOT

 True --> False, False --> True

 항상 Boolean 값으로 값을 반환한다!

```python
# and
print(True and True)
print(True and False)
print(False and True)
print(False and False)

print(True and 3)       # 3
print(3 and 5)         # 5 = 마지막으로 평가된 피연산자.
print(0 and 5)         # 0

a, b = 10, 5
result1 = a > b #True
result2 = a == b #False
print(result1 and result2)

# or
print(True or True)
print(True or False)
print(False or True)
print(False or False)

print(False or 3)      # 3
print(3 or 5)          # 3 왜? or에서 3이 True면 뒤는 볼 필요도 없으니까!
print(0 or 5)          # 5

# not
print(not True)        # False
print(not 0)           # True
print(not 3)           # False
```

```python
print(True or True and False) #?
print(4 or 2 and 0) #얘 왜 거꾸로 가지?
# 이유! OR 연산자가 True가 나오면서 AND 연산을 하기전에 연산을 끝내기 때문.
print(True or True and True)
print(True and True or False)
print(True and True or True)
print()

print(True or False and False) #?
print(True or False and True)
print(True and False or False)
print(True and False or True)
print()

print(False or True and False) #????? 얘는 왜 AND까지 하짐? 아!
# or 연산자에서 앞의 인수가 True면 무조건 True이다! 라는 관념으로 인해
#효율과 최적화의 측면에서 연산을 줄이기 위해 뒤의 연산을 하지 않는다.
#만약, OR를 쓸 일이 있다면, 첫 인수를 주의해서 사용하자.

print(False or True and True)
print(False and True or False)
print(False and True or True)
print()

print(False or False and False)
print(False or False and True)
print(False and False or False)
print(False and False or True)
print()
```
