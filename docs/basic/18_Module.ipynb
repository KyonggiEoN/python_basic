**1. 코드를 math_tools.py 로 저장**

````python
# 변수
PI = 3.141592653589793

# 함수
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

# 클래스
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return PI * self.radius * self.radius
````

# **3. 패키지**

파이썬에서 패키지는 모듈들을 포함하고 있는 디렉터리.

모듈은 파이썬 코드를 담고 있는 파일이며, 여러 모듈을 논리적으로 그룹화하려는 경우 패키지를 사용함.

패키지를 사용하면 관련된 기능들을 함께 묶어서 코드를 더욱 체계적으로 관리할 수 있다.

```
my_package/
│
├── __init__.py
│
├── module_a.py
│
└── sub_package/
    │
    ├── __init__.py
    └── module_b.py
```

패키지는 기본적으로 디렉터리와 그 안의 .py 파일들로 구성.

패키지를 구성하는 디렉터리 내에는 __init__.py 파일이 있어야 한다.. 이 파일은 해당 디렉터리가 파이썬 패키지임을 나타냄.

__init__.py는 비워둘 수도 있고, 패키지 초기화 코드를 포함할 수도 있다. 또한 패키지 안에 다른 패키지(하위 패키지)를 포함할 수 있다.

이를 통해 복잡한 프로젝트의 코드를 여러 레벨의 디렉터리로 계층적으로 구성할 수 있다.

```
from my_package import module_a
from my_package.sub_package import module_b
```

# **4. 패키지 실습**

```
shapes/
│
├── __init__.py
├── circle.py
└── rectangle.py
```

패키지의 초기화 코드를 포함할 수 있다.

예를 들어, 패키지가 import 될 때 특정 변수를 설정하거나 다른 초기화 작업을 수행하면 __init__.py 파일에 해당 코드를 작성함.

__init__.py 파일을 사용해 패키지 레벨에서 import를 제어할 수 있다.

예를 들어, from shapes import *를 했을 때 어떤 모듈을 import 할 지 정한다.

아래 설정은 from shapes import *를 사용할 때 circle와 rectangle 모듈만 가져옴

````python
'''
__all__ = ['circle', 'rectangle']
'''
````

````python
import shapes

print(shapes.circle.PI)

area = shapes.circle.circumference(10)
print(area)

recA = shapes.rectangle.area(10,23)
print(recA)

recP = shapes.rectangle.perimeter(10,23)
print(recP)
````

패키지 속 두 개의 파일을 불러왔음을 확인 할 수 있다.

````python
from shapes.circle import area as cA
from shapes.rectangle import area as rA

print(cA(3))
print(rA(10, 23))
````

모듈 속의 특정 함수만 불러올 수도 있다.
