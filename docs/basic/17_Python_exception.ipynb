# **2. 발생하는 경우**

ValueError

  * 잘못된 값을 함수나 연산에 제공할 때 발생합니다.
예) 숫자가 아닌 문자열을 int() 함수로 변환하려고 할 때 발생.

TypeError

  * 올바르지 않은 유형의 객체를 연산에 사용하려 할 때 발생합니다.
예) 문자열과 숫자를 함께 더하려고 할 때 발생.

ZeroDivisionError

  * 숫자를 0으로 나누려고 할 때 발생합니다.

IndexError

  * 리스트, 튜플, 문자열 등의 시퀀스 유형에서 범위를 벗어난 인덱스에 접근하려 할 때 발생합니다.
예) 길이가 3인 리스트에 대해 4번째 요소에 접근하려고 할 때 발생.

KeyError

  * 딕셔너리에서 존재하지 않는 키를 사용하여 값을 검색하려고 할 때 발생합니다.

AttributeError

  * 객체에 없는 속성이나 메서드에 접근하려고 할 때 발생합니다.

FileNotFoundError

  * 존재하지 않는 파일을 열려고 할 때 발생합니다.

ImportError

  * 존재하지 않는 모듈을 가져오려고 할 때 또는 모듈 내에 해당 속성/함수가 없을 때 발생합니다.

NameError

  * 정의되지 않은 변수나 함수를 사용하려고 할 때 발생합니다.
예) 프로그램에서 정의되지 않은 변수 x를 사용하려고 할 때 발생.

OverflowError

  * 수치 연산 결과가 너무 커서 표현할 수 없을 때 발생합니다.

MemoryError

  * 프로그램이 사용 가능한 모든 메모리를 소진했을 때 발생합니다.

# **4. Exception Class**

사용자 예외를 만들거나 특정 예외 유형을 잡기 위한 기본 인터페이스 제공.

예외 유형은 Exception을 상속받아서 정의됨.

 ValueError, TypeError, FileNotFoundError 등...





```python
try:
  #예외를 발생시키는 연산자 raise
  raise Exception('예외 발생')
except Exception as e:
  print(e)
```

```python
def func1():
    n = int(input('짝수를 입력하세요: '))
    if n % 2 == 1:
        raise Exception('짝수를 입력해주세요.')
    print(n)

# func1()

try:
    func1()
except Exception as e:
    print('예외가 발생: ', e)
```

```python
def func1():
  try:
    func2()

  except TypeError as e:
    print('예외 발생1')

  except:
    print('예외 발생2')

def func2():
  func3()

def func3():
  try:
    print('%d' % '문자열' )
  except TypeError:
    print('예외 발생3')
  raise Exception('예외 발생4')

func1()
```

# **5. 사용자 정의 예외 클래스**

```python
class AgeLimitError(Exception):
  def __init__(self, age, message='Error: Out of Range'):
    self.age = age
    self.message = message

    super().__init__(self.message)
    # Exception 객체를 만들면서 부모 생성자에 메세지를 보냄.

def checkAge(age):
  if age < 14:
    raise AgeLimitError(age, 'down')
  elif age > 43:
    raise AgeLimitError(age, 'up')
    # 이때 에러 객체를 생성함.
    # 객체는 변수 없이 만들어 질 수 있다.
    # 1회성으로 만들고 나중에 덮어쓴다.
    # AgeLimitError = raise AgeLimitError(age, ... )

  else:
    return f'{age}; in'

ages = [17, 60, 46, 20, 80, 55, 26, 12, 5]

for a in ages:
  try:
    print(checkAge(a))
  except AgeLimitError as e: #생성된 객체를 e로 Alias
    print(f'{e.age}; Age Error: {e}')
```
