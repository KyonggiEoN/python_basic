## ÏßëÌï© Ïó∞ÏÇ∞

```python
s3 = {10, 20, 30, 40, 50}
s4 = {30, 40, 50, 60, 70}

# union(): Ìï©ÏßëÌï©ÏùÑ Í≥ÑÏÇ∞ÌïòÏó¨ Î∞òÌôò
result1 = s3.union(s4)
print(result1)

result2 = s3 | s4
print(result2)
print(sorted(result2)) #sorted() Ìï®ÏàòÎäî Í∞íÏùÑ Î¶¨Ïä§Ìä∏Î°ú Î∞òÌôòÌï®Ïóê Ïú†ÏùòÌïòÏûê.0

# intersection(): ÍµêÏßëÌï©ÏùÑ Í≥ÑÏÇ∞ÌïòÏó¨ Î∞òÌôò
result1 = s3.intersection(s4)
# result1.sort()
#sorted() Ìï®ÏàòÎäî ÏßÄÏõêÌïòÏßÄÎßå sort Î©îÏÑúÎìúÎäî ÏßÄÏõêÌïòÏßÄ ÏïäÏùå
print(result1)

result2 = s3 & s4
print(result2)

# difference(): Ï∞®ÏßëÌï©ÏùÑ Í≥ÑÏÇ∞ÌïòÏó¨ Î∞òÌôò
result1 = s3.difference(s4)
print(result1)

result2 = s3 - s4
print(result2)

# symmetric_difference(): ÎåÄÏπ≠ Ï∞®ÏßëÌï©ÏùÑ Í≥ÑÏÇ∞ÌïòÏó¨ Î∞òÌôò
#ÎåÄÏπ≠ Ï∞®ÏßëÌï©? Ìï©ÏßëÌï© - ÍµêÏßëÌï©
result1 = s3.symmetric_difference(s4)
print(result1)

result2 = s3 ^ s4
print(result2)
```
