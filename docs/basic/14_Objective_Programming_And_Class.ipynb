# **1. 프로그래밍 방법론**

프로그래밍 방법론은 프로그램을 개발하는 다양한 접근 방식, 철학을 뜻함.

프로젝트 규모, 요구 사항, 개발팀 특성에 따라 적합한 방법론이 달라지므로 각 방법론의 특징과 장단점을 이해하는게 중요.

## 1-1. 절차적 프로그래밍(Procedural Programming)

 절차(Procedure)를 중시하는 방식으로, 작업을 순서대로 실행하도록 프로그램을 구성합니다.

코드가 명령문으로 이루어져 있고, 작업 순서가 중요합니다.

변수와 함수가 함께 사용되며, 함수를 통해 프로그램의 기능을 분리하고, 코드 중복을 줄입니다.

## 1-2. 객체지향 프로그래밍 (Object-Oriented Programming, OOP)

절차(Procedure)를 중시하는 방식으로, 작업을 순서대로 실행하도록 프로그램을 구성합니다.

코드가 명령문으로 이루어져 있고, 작업 순서가 중요합니다.

변수와 함수가 함께 사용되며, 함수를 통해 프로그램의 기능을 분리하고, 코드 중복을 줄입니다.

## 1-3. 함수형 프로그래밍 (Functional Programming)

함수를 중심으로 프로그램을 구성합니다. 수학적인 함수 개념에 기초하고 있으며 코드가 간결하고 직관적입니다.

예를 들어 map, filter, reduce 등의 함수나 재귀를 주로 사용하여 데이터 처리를 간결하게 수행합니다.

# **2. 클래스 만들기**

class 키워드를 사용해 클래스를 정의함.

클래스 이름은 관례적으로 대문자 시작.

클래스 내부에는 속성(멤버 변수)와 메서드(멤버 함수)를 정의 할 수 있음.

```
class 클래스이름:
    # 클래스 속성(멤버 변수) 정의
    속성1 = 초기값1
    속성2 = 초기값2

    # 생성자 메서드 (생략 가능)
    def __init__(self, 매개변수1, 매개변수2, ...):
        # 인스턴스 속성 초기화
        self.속성1 = 매개변수1
        self.속성2 = 매개변수2

    # 메서드(멤버 함수) 정의
    def 메서드1(self, 매개변수1, 매개변수2, ...):
        # 메서드 동작 정의
        pass

    def 메서드2(self, 매개변수1, 매개변수2, ...):
        # 메서드 동작 정의
        pass

```

**객체(Object)**: 객체는 현실 세계의 개체, 추상적 개념을 모델링 한 것.

자동차, 사람, 은행 계좌 등.

객체는 데이터(속성, 상태)와 메서드(동작, 함수)로 구성됨.

클래스에서 만들어진 실제 데이터(프로그램 내에서 사용된 것)

**클래스(Class)**: 클래스는 객체를 만들기 위한 템플릿, 설계도.

객체의 공통 속성과 동작을 정의하며 객체를 생성하는데 사용됨.

"자동차" 클래스는 모든 자동차 객체가 가져야하는 속성(색상, 속도)와 메서드(주행, 정지)를 정의할 수 있음.

**인스턴스(Instance)**: 클래스를 기반으로 실제로 생성된 객체.

클래스는 여러 인스턴스를 생성할 수 있다.

각 인스턴스는 독립적인 데이터와 메서드를 가짐.

*객체가 어떤 클래스에서 나왔는지에 초점을 맞출 때 쓰는 용어.*

**속성(Attributes, Fields)**: 객체는 데이터를 저장하기 위한 속성을 가짐.

이런 속성은 객체의 상태, 특성을 변수로 나타냄.

자동차 객체의 속성은 색상, 속도, 모델 등이 있다.

**메서드(Methods, Functions)**: 객체는 데이터를 조작하거나 특정 작업을 수행하기 위한 메서드를 포함함.

이런 메서드는 함수와 비슷하지만,객체의 상태에 접근하고 조작할 수 있음.

자동차 객체의 메서드는 주행, 정지, 경적 울리기 등이 있다.

## 2-1. 객체(인스턴스) 생성

## 2-2. 객체 속성 초기화

객체를 생성한 후에는 객체의 속성을 초기화 할 수 있다.

연산자를 이용, 객체의 속성에 접근해 값을 할당함. 또한 연산자를 이용해 객체의 메서드 호출가능.

````python
class Dog:
  name = ''
  age = 0
  family = ''
  def eat(self):
    #self는 생성된 속성을 가리킴
    print(f'{self.name}는 지금 밥 묵는 중...')

ari = Dog()
print(ari.name, ari.age, ari.family) #클래스의 default 값이 나온다.

ari.name = '아리'
ari.age = 15
ari.family = '요크셔테리어'

print(ari.name, ari.age, ari.family)
ari.eat()

mimi = Dog()
mimi.name = '미미'
mimi.age = 14
mimi.family = '말티즈'
print(mimi.name, mimi.age, mimi.family)
mimi.eat()
````

## 2-3. 생성자

생성자(Constructor)는 클래스의 인스턴스가 생성될 때 자동으로 호출되는 메서드.

생성자는 객체의 초기화를 담당하고, 객체가 생성될 때 필요한 속성을 초기화하고 설정하는 역할을 함.

생성자 메서드는 '\_\_init\_\_'으로 이름이 정해져 있음.

속성은 self를 사용해 접근, 설정함.

초기화된 속성은 객체가 생성될 때마다 다른 값을 가질 수 있음.

```
class 클래스이름:
    def __init__(self, 매개변수1, 매개변수2):
        self.속성1 = 매개변수1
        self.속성2 = 매개변수2
```

````python
class Dog:
    def __init__(self):
        print(self, 'init 호출!')

Rucy = Dog()
````

````python
class Dog:
    def __init__(self, name, age, family='족보 없음'):
        self.name = name
        self.age = age
        self.family = family

# Rucy = Dog()
# TypeError: Dog.__init__() missing 2 required positional arguments: 'name' and 'age'

ari = Dog('아리', 15, '요크셔테리어')
print(ari.name, ari.age, ari.family)

mimi = Dog('미미', 14)
print(mimi.name, mimi.age, mimi.family)
````

생성된 각 속성들은 같은 클래스의 같은 메서드, 변수를 쓰지만 고유의 값을 가진다.

````python
class Counter:
    def __init__(self):
        self.num = 0
    def increment(self):
        self.num += 1
    def current_value(self):
        return self.num
    def decrement(self):
        self.num -= 1
    def reset(self):
        self.num = 0

KBBank = Counter()
print(KBBank.num)
print(KBBank.current_value())
print()

KBBank.increment()
KBBank.increment()
KBBank.increment()
print(KBBank.current_value())
KBBank.decrement()
print(KBBank.current_value())
KBBank.reset()
print(KBBank.current_value())
for i in range(5): KBBank.increment()
print()

HanaBank = Counter()
print(HanaBank.current_value())
HanaBank.increment()
HanaBank.increment()
HanaBank.increment()
HanaBank.increment()
for i in range(5): HanaBank.increment()
print(KBBank.current_value())
print(HanaBank.current_value())
````

## 2-4. 클래스의 변수와 인스턴스 변수

***클래스 변수*** 는 클래스 자체에 속해 모든 인스턴스가 공유하는 변수로,

한 번 변경하면 모든 인스턴스에 영향을 미친다.
__________
***인스턴스 변수*** 는 각 인스턴스(객체)마다 별도로 존재하는 변수로,

인스턴스마다 독립적인 값을 가질 수 있어 다른 인스턴스에 영향을 주지 않는다.


````python
class nuum:
  num1 = 0 # 클래스 변수
  def __init__(self):
    self.num2 = 0 # 인스턴스 변수
  def pNum(self):
    return  nuum.num1, self.num2
  def num1up(self):
    nuum.num1 += 1
  def num2up(self):
    self.num2 += 1

a = nuum()
b = nuum()

for i in range(1): a.num1up()
for i in range(2): a.num2up()
for i in range(3): b.num1up()
for i in range(5): b.num2up()

print(a.pNum())
print(b.pNum())
````

````python
class Dog:
    family = "포메" # 클래스 변수

    def __init__(self, name, age):
        # 인스턴스 변수
        self.name = name
        self.age = age

    def print_info(self):
        print(f'종: {Dog.family}')  # 클래스 변수
        print(f'이름: {self.name}') # 인스턴스 변수
        print(f'나이: {self.age}')  # 인스턴스 변수

Rucy = Dog('루시', 15)
PPomi = Dog('뽀미', 7)
Rucy.print_info()
PPomi.print_info()
print()

Dog.family = '폼피츠'
Rucy.print_info()
PPomi.print_info()
print()

Rucy.family = '포메'
# Rucy의 것을 바꾸어도 클래스 변수는 변함없다.
Rucy.print_info()
PPomi.print_info()
print()

print(Rucy.family)
print(PPomi.family)
# 클래스 변수와 다르게 인스턴스 변수는 보존됨을 알 수 있다.
````

## 2-5. 메서드 타입

***인스턴스 메서드(Instance Method)*** : 객체의 상태를 조작하거나 객체에 특정 작업을 수행하는 메서드.

대부분의 클래스 속 메서드는 인스턴스 메서드. \_\_init\_\_ 도 인스턴스 메서드.
____________
***클래스 메서드(Class Method)*** : 클래스 레벨에서 동작하며, 모든 인스턴스가 공유하는 메서드.

클래스 메서드는 @classmethod 데코레이터를 사용하여 정의하며, 첫 번째 매개변수로 cls를 사용함.

클래스 변수를 가져다 쓸 수 있다.
____________
***정적 메서드(Static Method)*** : 특정 클래스나 인스턴스와 관련이 없는 메서드

클래스 내부에 정의되어 있지만 클래스나 인스턴스와 독립적으로 호출될 수 있음.

정적 메서드는 @staticmethod 데코레이터를 사용하여 정의.

````python
class Calculator:
    def add(self, x, y):
        return x + y

    @classmethod
    def subtract(cls, x, y):
        return x - y

    @staticmethod
    #self나 cls를 받지 않는다.
    def multiply(x, y):
        return x * y

# 인스턴스 생성
calc = Calculator()

# 인스턴스 메서드 호출
result1 = calc.add(5, 3)
print(result1)  # 출력: 8

# 클래스 메서드 호출
result2 = Calculator.subtract(10, 4)
result22 = calc.subtract(10, 4)
print(result2, result22)  # 출력: 6

# 정적 메서드 호출
result3 = Calculator.multiply(6, 2)
print(result3)  # 출력: 12
````
