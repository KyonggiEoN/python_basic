# **1. 스페셜 메서드**

더블 언더스코어(\_\_)로 시작하고 끝나는 메서드.

특정 구문이나 내장 함수를 사용할 때 파이썬 인터프리터에 의해 자동으로 호출됨.

## 1. \_\_repr\_\_()

**공식적인** 문자열 표현을 정의하는 메서드

주로 개발자나 디버깅을 위한 정보를 제공.

객체를 문자열로 변환할 때 자동으로 호출됨.

```python
class Dog:
  def __init__(self, name, age):
    self.name = name
    self.age = age

ari = Dog('아리', 15)
print(repr(ari))
print(ari)
```

```python
class Dog:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def __repr__(self):
    return f"Dog(name= '{self.name}', age= {self.age})"

ari = Dog('아리', 15)
print(repr(ari))
print(ari)
```

개체의 주소 값이 아니라 지정한 값을 반환함.

```python
# eval()은 주어진 문자열을 파이썬 표현식으로 평가하고 실행하여 그 결과를 반환하는 내장 함수

x = 10
y = 3
result = x + y
print(result)
result = eval("x + y")
print(result)

Rucy_repr = repr(ari)
result = eval(Rucy_repr)

print(result)
print(result == ari) # False, 같은 값을 가진 다른 객체
```

## 2. \_\_str\_\_()

사용자가 이해하기 쉽고 읽기 좋은 형태의 문자열 표현을 반환하는 메서드. 주로 객체 출력시 사용.

print()나 str()을 통해 호출됨.

\_\_repr\_\_()는 객체의 공식적인 표현을 제공하고, repr() 함수를 통해 호출됨.

```python
class Book:
    def __init__(self, title):
        self.title = title

book = Book('미친듯이 재밌는 파이썬')
print(book)
print(str(book))

class Book:
    def __init__(self, title):
        self.title = title
    def __str__(self):
        return self.title

book = Book('미친듯이 재밌는 파이썬')
print(book)
print(str(book))
print(repr(book))
```

## 3. \_\_len\_\_()

객체의 길이 또는 크기를 반환하는 메서드.

len() 함수가 호출될 때 자동으로 호출됨.

객체의 항목 수를 측정하거나, 특정 크기(리스트, 문자열, 튜플 등)을 나타내고자 할 때 구현됨.

```python
class Queue:
    def __init__(self):
        self.items = [1, 2, 3, 4, 5]

    def __len__(self):
        return len(self.items)

li = [1, 2, 3, 4, 5]
print(len(li))
print(li)

queue = Queue()
print(queue)
print(len(queue))
```

## 4. \_\_getitem\_\_()

인덱싱을 지원하기 위해 사용.

객체의 특정 항목을 가져오기 위해 대괄호를 쓸 때 호출됨.

객체의 특정 인덱스나 키에 해당하는 값을 반환하는 역할.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    # index 매개변수로 [] 속 숫자(0, 1, -100)를 받는다.
    def __getitem__(self, index):
        if index == 0:
            return self.x
        elif index == 1:
            return self.y
        else:
            return -1

pt = Point(5, 3)
print(pt)
print(pt[0])
print(pt[1])
print(pt[-100])
```

```python
class MyList:
    def __init__(self, data):
        self.data = data
    def __getitem__(self, index):
        return self.data[index]

ml = MyList([10, 20, 30, 40])
print(ml[0])
print(ml[1])
print(ml[-1])
```

본래 객체는 그 값을 주소형태(ex[<__main__.Point object at 0x798e4c3e3580>]) 로 가지고 있기에 [ ]나 len 등의 인덱싱 관련이 안 된다.

하지만 스페셜 메소드를 통해 오버라이딩을 하면, 이를 원하는 값을 주도록 사용할 수 있다.

## 5. \_\_call\_\_()

파이썬에서 객체를 함수처럼 호출 할 수 있게 해주는 헤서드.

클래스의 인스턴스를 함수처럼 사용할 수 있고, 인스턴스에 대해 소고라호를 사용해 값을 전달하고 결과를 반환받을 수 있음.

```python
class CallableObject:
    def __call__(self, *args, **kwargs):
        print(f'args:{args}, kwargs:{kwargs}')

# * 하나는 튜플로 묶어서, 두개는 키, 값의 딕셔너리로.
callable_obj = CallableObject()
callable_obj(1, 2, 3, a='A', b='B') # call 오버라이딩 효과
```
